Reshape ma tran

```python
# @title
arr_input = np.array([[1,2,3], [4,5,6]])
print("Truoc khi reshape")
print(arr_input)
print(arr_input.shape)
arr_input = arr_input.reshape(3, 2)
print("\nSau khi reshape")
print(arr_input)
print(arr_input.shape)
```

Flatten ma tran

```python
# @title
arr_input = np.array([[1,2,3], [4,5,6]])
print("\nFlatten mang 2 chieu")
arr_input = arr_input.flatten()
print(arr_input)
print(arr_input.shape)
print("\nFlatten mang 3 chieu")
arr_input = np.array([[[1,2,3], [4,5,6]],
                      [[1,2,3], [4,5,6]]])
arr_input = arr_input.flatten()
print(arr_input)
print(arr_input.shape)
```

Hstack va Vstack

```python
# @title
arr_input1 = np.array([1, 2, 3])
arr_input2 = np.array([4, 5, 6])

h_stack = np.hstack((arr_input1, arr_input2))
v_stack = np.vstack((arr_input1, arr_input2))

print("Horizon append:\n", h_stack)
print("Vertical append:\n", v_stack)
```

8. Viết chương trình tạo mảng 20 phần tử trong khoảng [0, 20] và đổi dấu những số nằm trong
khoảng [9, 15]

```python
# @title
import numpy as np

arr_input = np.random.randint(0, 20, size=10)
arr_condition = np.arange(9, 15)

condition = np.isin(arr_input, arr_condition)
result = np.where(condition, arr_input, -arr_input)

print(result)
```
